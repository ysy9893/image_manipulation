```python
import os 
import cv2
```

```python
#Extract only desirable annotations 
BASE_DIR='C:/Users/gptjd/OneDrive/Desktop/tflite_data/car/Label'
IMAGE_DIR='C:/Users/gptjd/OneDrive/Desktop/tflite_data/car/Image'
label_list=os.listdir(BASE_DIR)
image_list=os.listdir(IMAGE_DIR)
count=0
for label in label_list:
    
    count+=1
    f=open(os.path.join(BASE_DIR,label),'r')
    bbox=[]
    while 1:
        line=f.readline()
        if line:
            bbox.append(line.split())
        else:
            break 
        
    f.close()
    image=label.replace('.txt','.jpg')
    image=cv2.imread(os.path.join(IMAGE_DIR,image))
    cv2.imwrite('C:/Users/gptjd/OneDrive/Desktop/tflite_data/car/temp_image'+'/car'+str(count)+'.jpg',image)
    
    flag=0
    for b in bbox:
        if b[0]=='5.0':
            if flag==0:
                f=open('C:/Users/gptjd/OneDrive/Desktop/tflite_data/car/temp_label'+'/car'+str(count)+'.txt','w')
                flag=1
            text='car'+' '+b[1]+' '+b[2]+' '+b[3]+' '+b[4]
            f.write(text+'\n')
    if flag==0:
        print(label)
    


            
```

```python
##Extract only desirable line in txt file 

#첫번째, train.txt파일 정리 -> 각 이미지에 대한 annotation txt file을 만든다. 
#포맷은 class x1,y1,x2,y2 로!!
#파일 만든 후 각각 알맞은 이미지의 폴더에 저장한다. 

BASE_IMAGE_DIR='C:/Users/gptjd/OneDrive/Desktop/tflite_data/human/Image/images'
LABEL_DIR='C:/Users/gptjd/OneDrive/Desktop/tflite_data/human/train.txt'

f=open(LABEL_DIR,'r')
while True: 
    pick=[]
    image_dir=f.readline()#이미지 파일 디렉토리를 저장한다. 
    image_dir=image_dir.replace('.jpg','.txt').strip()
    if image_dir:
        count=f.readline().strip()
        count=int(count)
        #count가 0인 경우에는 바운딩 박스가 이미지에 없다는 것을
        #의미하므로 다음건너뛴다. 
        if count==0:
            f.readline()
            continue 
        for i in range(1,count+1):
            pick.append(f.readline().split()[:4])
        
    else:
        break

    n=open('C:/Users/gptjd/OneDrive/Desktop/tflite_data/human/Image/images/'+image_dir,'w')
    
    pick=convert(pick)
    
    for p in pick:
        x1=p[0]
        y1=p[1]
        x2=p[2]
        y2=p[3]
        text='human'+' '+str(x1)+' '+str(y1)+' '+str(x2)+' '+str(y2)+'\n'
        n.write(text)
    n.close()
f.close()
    
        
    
    
            
        
        
        
        
    
```

```python
def convert(pick):#x1,y1,w,h->x1,y1,x2,y2
    box=[]
    for p in pick:
        x1=int(p[0])
        y1=int(p[1])
        x2=int(p[0])+int(p[2])
        y2=int(p[1])+int(p[3])
        box.append([x1,y1,x2,y2])
    return box
        
```
