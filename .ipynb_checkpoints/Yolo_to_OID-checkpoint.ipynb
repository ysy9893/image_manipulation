```python
import os 
import cv2
```

```python
LABEL_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/Label'
IMAGE_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/Image'
NEW_LABEL_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/new_Label'
NEW_IMAGE_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/new_Image'

label_list=os.listdir(LABEL_PATH)
i=0
for file in label_list:
    i+=1
    image=file.replace('.txt','.jpg')
    box=[]
    f=open(LABEL_PATH+'/'+file,'r')
    
    while 1:
        line=f.readline()
        if not line:
            break
        box.append(line.split())
    f.close()
    
    
    
    image=cv2.imread(IMAGE_PATH+'/'+image)
    imH,imW=image.shape[:2]
    
    f=open(NEW_LABEL_PATH+'/'+'motorcycle'+str(i)+'.txt','w')
    
    for b in box:
        pick=[float(i) for i in b]
        x1,y1,x2,y2=convert(imH,imW,pick)
        
        text="0"+" "+str(x1)+" "+str(y1)+" "+str(x2)+" "+str(y2)
        print(text)
        f.write(text+'\n')
    f.close()
    
    
        
        
        
```

```python
def convert(imH,imW,pick):
    x=pick[1]*imW
    y=pick[2]*imH
    w=pick[3]*imW
    h=pick[4]*imH
    
    w_half=w*0.5
    h_half=h*0.5
    
    x1=int(x-w_half)
    y1=int(y-h_half)
    x2=int(x+w_half)
    y2=int(y+h_half)
    
    return x1,y1,x2,y2
    
```
