```python
import json 
import os
with open("C:/Users/gptjd/OneDrive/Desktop/tflite_data/car_train.json") as new_json:
    temp=json.load(new_json)
```

```python
print(temp[7])
```

```python
TXT_PATH='/content/drive/MyDrive/인턴프로젝트/tensorflow1/models/research/object_detection/images/txt'
i=0
count=0
image_list=[]
flag=0

for tp in temp:
    flag=0
    print(i)

    txt_name=tp['name'].replace('.jpg','.txt')

    f=open(os.path.join(TXT_PATH,txt_name),'w')

    if (tp['labels']!=None):
        elements=tp['labels']
        for element in elements:
            ec=element['category']
            if ec!='car' and ec!='truck' and ec!='bus': 
                continue
            else:
              flag=1
            category='car'
            x1=str(element['box2d']['x1'])
            y1=str(element['box2d']['y1'])
            x2=element['box2d']['x2']
            y2=element['box2d']['y2']
            text=category+" "+str(x1)+" "+str(y1)+" "+str(x2)+" "+str(y2)+'\n'
            f.write(text)
    f.close()
    if flag==1:
      image_list.append(tp['name'])
      count+=1
    if count==30000:
        break
    i+=1

```
