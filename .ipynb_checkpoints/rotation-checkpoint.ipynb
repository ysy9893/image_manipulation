2. 이미지를 회전시키고, 바운딩 박스도 그에 맞게 변환한다.

3. 이미지 회전 시 바운딩 박스 정보 변환 함수를 만든다. 

```python
def convert_bbox(x,y,num,imW,imH):
    #바운딩 박스의 중점 좌표를 구한다. 
    
    cx=x
    cy=y
    
    imW=imW/2
    imH=imH/2
    #이미지 프레임의 중점 좌표가 (0,0)이라고 했을 때, 
    #바운딩 박스의 중점 좌표 값도 이에 따라서 변환되어야 한다.
    
    #1. (+, +)
    if (cx>imW and cy<imH):
        cx=cx-imW
        cy=imH-cy
    #2. (+, -)
    elif (cx>imW and cy>imH):
        cx=cx-imW
        cy=-(cy-imH)
    #3. (-, -)
    elif (cx<imW and cy>imH):
        cx=-(imW-cx)
        cy=-(cy-imH)
    elif (cx<imW and cy<imH):
        cx=-(imW-cx)
        cy=imH-cy 
    else:
        cx=0
        cy=0
    #cos,sin값을 계산한다. 
    cos_1=math.cos(num)
    sin_1=math.sin(num)
    
    #새로운 좌표 값을 계산한다. 
    cx=cos_1*cx-sin_1*cy
    cy=sin_1*cx+cos_1*cy
    
    
    
    #이미지 크기에 알맞은 좌표 값으로 변환한다. 
    cx=imW+cx
    cy=imH-cy
    
    
    return cx,cy
    
    
```
