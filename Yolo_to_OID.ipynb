```python
def convert(imH,imW,pick):
    x=pick[0]*imW
    y=pick[1]*imH
    w=pick[2]*imW
    h=pick[3]*imH
    
    w_half=w*0.5
    h_half=h*0.5
    
    x1=x-w_half
    y1=y-h_half
    x2=x+w_half
    y2=y+h_half
    
    return x1,y1,x2,y2
    
```
