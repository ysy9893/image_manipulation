```python
import os

class_label={'Chair':'0','Light bulb':'1','Window':'2','Person':'3',
            'Door':'4','Picture frame':'5','Book':'6','Table':'7',
            'Drawer':'8','Houseplant':'9'}
```

```python
PATH='C:/Users/gptjd/OneDrive/Desktop/label'
txt_list=os.listdir(PATH)

for i in txt_list:
    f=open(PATH+'/'+i,'r')
    list_box=[]
    while 1:
        line=f.readline()
        if not line:
            break
        #    
        if (line.split()[0]=='Light' or line.split()[0]=='Picture'):
            concat=line.split()[0]+' '+line.split()[1]
            temp_list=[concat,line.split()[2],line.split()[3]
                      ,line.split()[4],line.split()[5]]
            list_box.append(temp_list)
            continue
                       
        else:
            list_box.append(line.split())
    f.close()
    
    print(list_box)
    print(i)
    f=open('C:/Users/gptjd/OneDrive/Desktop/label/'+i,'w')
    
    for j in list_box:
        text=class_label.get(j[0])+" "+j[1]+" "+j[2]+" "+j[3]+" "+j[4]
        f.write(text+'\n')
    f.close()
```
