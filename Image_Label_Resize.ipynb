```python
import os 
import cv2
```

```python
IMAGE_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/new_Image'
LABEL_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/new_Label'
RESIZED_IMAGE_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/resized_Image'
RESIZED_LABEL_PATH='C:/Users/gptjd/OneDrive/Desktop/tflite_data/motorcycle/resized_Label'

image_list=os.listdir(IMAGE_PATH)

for image in image_list:
    if image=="desktop.ini":
        continue
    name=image
    image=cv2.imread(IMAGE_PATH+'/'+image)
    print("image.shape={0}".format(image.shape))
    
    imH,imW=image.shape[:2]
    
    image=cv2.resize(image,(300,300))
    print("new_image.shape={0}".format(image.shape))
    
    cv2.imwrite(RESIZED_IMAGE_PATH+'/'+name,image)
    
    
    label=name.replace('.jpg','.txt')
    
    f=open(LABEL_PATH+'/'+label,'r')
    
    box=[]
    while 1:
        
        line=f.readline()
        if not line:
            break
        box.append(line.split())
    f.close()
    
    f=open(RESIZED_LABEL_PATH+'/'+label,'w')
    
    for b in box:
        pick=[float(b[i]) for i in range(1,len(b))]
        
        x1=pick[0]*(300.0/imW)
        y1=pick[1]*(300.0/imH)
        x2=pick[2]*(300.0/imW)
        y2=pick[3]*(300.0/imH)
        
        text='motorcycle'+" "+str(x1)+" "+str(y1)+" "+str(x2)+" "+str(y2)
        f.write(text+"\n")
    f.close()
        
        
        
    
    
    
    










```
